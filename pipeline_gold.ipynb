```python
import pyspark
import pyspark.sql
import pyspark.sql.functions as F
from pyspark.sql.functions import col

spark = spark.builder.getOrCreate()
```

```python
df_silver = spark.read.table("default.sncf_silver_clean")

df_silver.show(5)
```

```python
#Table de dimension gare (Dim_Gare)

gare_depart = df_silver.select(col("gare_de_depart").alias("nom_gare"))
gare_arrivee = df_silver.select(col("gare_d_arrivee").alias("nom_gare"))

df_dim_gare = gare_depart.union(gare_arrivee).distinct()

df_dim_gare = df_dim_gare.withColumn("gareID", F.monotonically_increasing_id())

df_dim_gare.show(5, truncate = False)
```
