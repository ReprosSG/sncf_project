```python
import pyspark
import pyspark.sql
import pyspark.sql.functions as F
from pyspark.sql.functions import col

spark = spark.builder.getOrCreate()
```

```python
df_silver = spark.read.table("default.sncf_silver_clean")

df_silver.show(5)
```

```python
#Table de dimension gare (Dim_Gare)

gare_depart = df_silver.select(col("gare_de_depart").alias("nom_gare"))
gare_arrivee = df_silver.select(col("gare_d_arrivee").alias("nom_gare"))

df_dim_gare = gare_depart.union(gare_arrivee).distinct()

df_dim_gare = df_dim_gare.withColumn("gareID", F.monotonically_increasing_id())

df_dim_gare.show(5, truncate = False)
```

```python
#Table de fait

df_fait_regularite = (df_silver
                      .groupBy("date", "service", "gare_de_depart", "gare_d_arrivee")
                      .agg(F.sum("nombre_de_trains_en_retard_a_l_arrivee").alias("total_trains_en_retard"),
                           F.sum("nombre_de_circulations_prevues").alias("total_trains_programmes")))

df_fait_regularite = df_fait_regularite.withColumn("taux_regularite", F.when(col("total_trains_programmes") > 0, (1 - (col("total_trains_en_retard")/col("total_trains_programmes")) * 100)).otherwise(100))

df_fait_regularite = df_fait_regularite.withColumn("taux_regularite", F.round(col("taux_regularite"), 2))
```

```python
#Ajout dans le catalogue

df_dim_gare.write.mode("overwrite").saveAsTable("default.dim_gare")

df_fait_regularite.write.mode("overwrite").saveAsTable("default.fait_regularite")

print("Bien deployee dans Unity Catalogue")
```
